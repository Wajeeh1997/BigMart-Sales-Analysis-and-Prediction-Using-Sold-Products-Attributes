```python
#lable encoding

le = LabelEncoder()
Label = ['Item_Fat_Content','Outlet_Size','Outlet_Location_Type']

for i in Label:
    Bigmart_tr[i] = le.fit_transform(Bigmart_tr[i])
    Bigmart_te[i] = le.fit_transform(Bigmart_te[i])

Bigmart_tr.head()
```

```python
#one hot encoding
cols = ['Item_Type','Outlet_Type']
# Apply one-hot encoder
OH_encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)
tr_oh = pd.DataFrame(OH_encoder.fit_transform(Bigmart_tr[cols])).astype('int64')
te_oh = pd.DataFrame(OH_encoder.fit_transform(Bigmart_te[cols])).astype('int64')


# #get feature columns
# tr_oh.columns = OH_encoder.get_feature_names(cols)
# te_oh.columns = OH_encoder.get_feature_names(cols)

# One-hot encoding removed index; put it back
tr_oh.index = Bigmart_tr.index
te_oh.index = Bigmart_te.index

# Add one-hot encoded columns to our main df new name: tr_fe, te_fe (means feature engeenired)
tr_fe = pd.concat([Bigmart_tr, tr_oh], axis=1)
te_fe = pd.concat([Bigmart_te, te_oh], axis=1)
```
